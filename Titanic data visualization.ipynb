```python
# Splitting "Name" column into first name and last name
tit['Name'] = tit['Name'].str.split(',')
tit['First Name'] = tit['Name'].str[1].str.strip()
tit['Last Name'] = tit['Name'].str[0].str.strip()

# Dropping the original "Name" column
tit.drop(columns=['Name'], inplace=True)

# Printing the updated DataFrame
print(tit)
```

```python
tit
```

```python
tit["Age"]=tit["Age"].fillna(np.mean(pd.to_numeric(tit["Age"])))
tit
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
tit['Age'] = scaler.fit_transform(tit[['Age']])
tit
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming you have a DataFrame named 'df' containing the data

# Extracting ticket prefix
tit['TicketPrefix'] = tit['Ticket'].str.extract('([A-Za-z\.]+)', expand=False)

# Handling missing values
tit['TicketPrefix'].fillna('Unknown', inplace=True)

# Encoding categorical values
label_encoder = LabelEncoder()
tit['TicketPrefix'] = label_encoder.fit_transform(tit['TicketPrefix'])

# Drop the original 'Ticket' column if not needed for the model
tit = tit.drop('Ticket', axis=1)
tit
```

```python
tit['Fare'] = tit['Fare'].round(2)
tit
```

```python
tit['Has_Cabin'] = np.where(tit['Cabin'].isnull(), 0, 1)
tit
```

```python
tit = tit.drop('Cabin', axis=1)
tit
```

```python
# Perform one-hot encoding on the 'Sex' column
tit_encoded = pd.get_dummies(tit['Sex'], prefix='Sex')

# Concatenate the encoded columns with the original DataFrame
tit = pd.concat([tit, tit_encoded], axis=1)

# Drop the original 'Sex' column
tit = tit.drop('Sex', axis=1)
tit
```
