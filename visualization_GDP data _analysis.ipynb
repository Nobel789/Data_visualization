```python
##### Visualization
# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

#import warnings
#warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
# see Distributions of attributes: univariation, you see the range o to 20 its not properly marked
vis1 = sns.displot(df["usersinternet"])
vis1
```

```python
#Univariation analysis with bins
#Binning data into intervals allows you to observe the frequency or count of values falling within each interval.
vis2 = sns.displot(df["usersinternet"],bins=10)
vis2
```

```python
#BOX PLOTS:
vis3 = sns.boxplot(data = df, x="usersinternet", y='Birthfrequency')
vis3
```

```python
vis4= sns.boxplot(data = df, x="incometype,", y='Birthfrequency')
vis4
```

```python
# Advanced visualization like linear model regressin  with seaborn, by default u get linear line
vis5= sns.lmplot(data = df, x="usersinternet", y='Birthfrequency')
vis5
```

```python
#without the linear line
vis6= sns.lmplot(data = df, x="usersinternet", y='Birthfrequency',fit_reg=False)
vis6
```

```python
#you can see in above lm plot the relationship or regression betwen userinternet and birthfrequency all are uniform
#To differentiate you can use "hue2 for income group , its differentiate the income group parameter and also show aligend
vis7 = sns.lmplot(data = df,x="usersinternet", y='Birthfrequency',fit_reg = False, hue = 'incometype,')
vis7
```

```python
## you see its marked and showing as legend in right side of bar
```

```python
#To see the large visualization, input the "height" parameter
vis7 = sns.lmplot(data = df,x="usersinternet", y='Birthfrequency',fit_reg = False, hue = 'incometype,',height=9,)
vis7
```
