# Extensive EDA and visualization

Cardiovascular disease, also known as heart disease, is a major cause of death worldwide, except in Africa. It accounted for a significant number of deaths in 2015. While rates of cardiovascular disease have been declining in developed countries, they have been increasing in many developing nations.

In this project, the author performed Exploratory Data Analysis (EDA) on a dataset related to heart disease. EDA is an essential initial step in analyzing a new dataset. Its main purpose is to examine the data for its distribution, identify outliers and anomalies, and gain insights that can guide further hypothesis testing. EDA involves studying the data to understand its distribution, characteristics, patterns, and visual representations. It helps generate hypotheses by providing visual tools and a better understanding of the data.
